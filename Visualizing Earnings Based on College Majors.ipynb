## Introduction

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
print(recent_grads.describe())
```

```python
raw_data_count = len(recent_grads)
recent_grads = recent_grads.dropna(axis = 0)
cleaned_data_count = len(recent_grads)
print(raw_data_count)
print(cleaned_data_count)
```

## Pandas, Scatter Plots

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title = "Median vs. Sample_size")
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title = "Unemployment_rate vs. Sample_size")
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title = "Median vs. Full_time")
```

### Findings
There doesn't seem to be any correlation between number of full time employees and their median salary.

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title = "Unemployment_rate vs. ShareWomen")
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title = "Median vs. Men")
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title = "Median vs. Women")
```

```python
recent_grads.plot(x='Total', y='Median', kind='scatter', title = "Median vs. Total")
```

### Findings
The plot above explains that the most popular majors do not actually make more money than average.

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title = "Median vs. ShareWomen")
```

### Findings
The plot above suggests that there is a weak correlation between share of women in a major and median income. This implies that the majors women mainly pursue do not actually help them make more money than average.
